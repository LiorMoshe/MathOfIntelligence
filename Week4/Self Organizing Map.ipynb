### Self Organizing Map

In this week's challenge we will implement a self organizing map,or **SOM** in short.
SOM is a type of artificial neural network that is used for unsupervised learning tasks in order to produce a **mapping** from the feature space of the input data to a low dimensional space. This algorithm is useful for visualizing high dimensional data in low-dimensional space.

The most popular neural networks train using an optimization algorithm like SGD where the learning of the network is achieved by learning from it's **errors**,this type of learning is often called error-correction learning. SOM uses a different technique called **competitive learning**.

The SOM network is usually visualized by a structure called the **SOM grid** , this structure is a grid that displays all the neurons in the network, each neuron is initialized by a weight vector of the space dimensionality of the input space. For example: If we have a 10 dimensional input data and 256 neurons we have 256*10 = 2560 weights.

#### The Algorithm

So how do we learn our mapping?<br>
As we said our training utilizes competitive learning.Given a matrix $X$ of dimensions $(n,m)$ (meaning that we have n training examples where each training example has m features). We will initialize our weight matrix $W$ to be of shape $(m,N)$ where N is the number of neurons in our network.

**Note**: From now on when I relate to a node in the map this is the same as a neuron (node=neuron).

##### Variables
$s$ = The current iteration. <br>
$t$ = The current index of the current training example in the training data. <br>
$v$ = Index of the current node in the map.<br>
$u$ = Index of the best matching unit in the map.<br>
$\theta(u,v,s)$ = The neighborhood function,restraint on different nodes in the map based on their distance from the BMU. <br>
$\alpha(s)$ = Simple to the learning rate in SGD,this time this is a function of the iteration that we are in, we want the learning rate to reduce in time (same as learning rate anealling).<br>

Then in order to train we pass through the following steps:<br>
1. Randomly pick a training example $X(t)$
2. Traverse each neuron in our mapping and do:<br>
<tab>2.1. Compute the euclidean distance between the input vector and the neuron's weights.<br>
<tab>2.2. Track the node that has the smallest distance, this node is called the **BMU** best<br> <tab> <tab>matching unit (this is the competitive part of our learning).
3. Update the weight vectors of the BMU and it's neighbors using the following update rule:
    $$W_v(s + 1) = W_v(s) + \theta(u,v,s) * \alpha(s) * (X(t) - W_v(s))$$ <br>
4. Increase s until we get to our iteration limit.


#### Exaplanation
Because we have to be a little bit different from wikipedia, let's explain what the hell is going on here. <br>
Before training our algorithm our mapping is somewhere along a m-dimensional space (which may be no where near our input data). Our final goal is that our map will find it's path to the training data in order to find clusters in this data. Great illustration of this process:
![title](750px-Somtraining.svg.png)

In this illustration we can see our grid of nodes (the white grid) and the input data (the blue blob). At the beggining before we even train we can see that they are no where near each other, in each training step they get closer and closer until at the end of the training our grid of nodes is all over the training data. <br>
We are achieving this by using our **update rule**, this is the scary equation that we saw in step 3 in our training algorithm.<br>
After finding the BMU we want to update the nodes in our map in order to represent the training data better. But in the same time we want to divide the data into different clusters. That means that if we change the weights of specific nodes in our map we will change the weights of the BMU and it's **neighbors**, we don't want to ruin what we already learned in the past about different clusters.<br>
So how can we define neighbors? <br>
Here the neighborhood function $\theta(u,v,s)$ comes into the scene. In it's simplest form we define some limit distance $d$ and our neighborhood function as :<br>
$$ 
  \begin{equation*}
  \theta(u,v,s)=\begin{cases}
    1, & |u - v| < d \\
    0, & otherwise \\\end{cases}\end{equation*}
$$

Meaning, if your distance from the BMU is smaller than $d$ you are considered a neighbor of the the BMU. <br>
The idea is that the neighborhood of each node in our map **decreases** over time, at the beggining it is broad and each training iteration has a global effect, but when we get further in training each node's neighborhood shrinks to the size of it's cluster and each training example have a local effect on the BMU's cluster. <br>

 After we realize the meaning of the neighborhood function everything gets clearer. What we do in the update rule is passing over all the nodes in our map (marked as $v$ in the current iteration). And update their weights using the neighborhood function. If this node is a neighbor of the BMU we update it's weights using it's weights distance from the random chosen training example $X(t)$ and the learning rate function $\alpha(t)$. The neighborhood's distance that we refered to as $d$ is often called the **radius** of our map.


Now that we understood how the algorithm works let's get to the fun part, the **code**:

### The data
Explain here about the dataset that we will use. (probably iris or something).

Let's implement our self organizing map as a class (over several cells).

```python
class SOM():
    def __init__(self,x_size,y_size,input_dim,max_iter):
        '''
        Constructor of our self organizing map.
        Parameters:
            - x_size,y_size : Dimensions of our SOM grid.
            - input_dim: The input dimension: This will be the dimension of the weights of
                         each one of the nodes in the grid.
            - max_iter: The maximum number of training iterations.
        '''
        #Take random integers up to 256 for the weights initialization.Change type to numpy floats
        self.weights = np.random.randint(256,size=(x_size,y_size,input_dim)).astype('float64')
        self.max_iter = max_iter
        
        #Initial radius for the BMU's neighborhood will be half the grid's size.
        self.radius = max(self.weights.shape) / 2
        
        #Compute the constant that will be used to update the radius.
        self.update_const = self.max_iter / math.log(self.radius)
```

In our constructor we initialized the weights for all the nodes in the grid (there are initializations that make our algorithm converge faster but I won't get into it in this notebook,for more about this use this [link](https://en.wikipedia.org/wiki/Self-organizing_map#SOM_Initialization).
Another parameters that we will save in our class is the max number of iterations that will be ran while training the network (named max_iter), the initial radius that will be used to find the neighborhood of the BMU and finally a constant that will be used while updating our radius (more about this soon).

Let's go on to the code that will compute the BMU given a random training example:

```python
class SOM(SOM):
    def compute_distances(self,trn):
        '''
        Compute the distances of all the weights of the nodes in our grid from the example
        training example trn.
        '''
        return np.sum((self.weights - trn) ** 2,2)
    
    def compute_bmu(self,trn):
        '''
        Compute the best matching unit given some random training example as input:trn.
        This will return the index of the BMU in our grid.
        '''
        distance_matrix = self.compute_distances(trn)
        #Unravel index receives index in flattened shape and converts to the matching index
        # in the given shape in the second input variable.
        return np.unravel_index(distance_matrix.argmin(),distance_matrix.shape)
    
    def compute_bmu_distances(self,trn):
        '''
        Compute the distances of all the nodes from the best matching unit of the given random
        training example trn.
        Note: This is the distance of the nodes on the grid, meaning we compute the distance
        between their indices in the matrix that represents the grid.
        '''
        x_size,y_size,input_dim = self.weights.shape
        
        #Create matrix of all indexes in both dimensions.
        x_idxs = np.arange(x_size).reshape(x_size,1).repeat(y_size,1)
        y_idxs = np.arange(y_size).reshape(1,y_size).repeat(x_size,0)
        
        #Use depth stack to get all available indexes and compute their distance from the bmu
        #in our grid.
        return np.sum((np.dstack((x_idxs,y_idxs)) - np.array(self.compute_bmu(trn))) ** 2,2)
    
```

The code above uses several useful numpy functions to compute the distance of all the weights from a given point and finally compute the best matching unit to a given training example (this is the competitive part of our training).

###### Updating the radius
As we said when we described the algorithm, the neighborhood of each node in our graph is going to shrink as we iterate and go through our training. In order to do it we need to make our radius shrink through training.<br>
In order to control our radius we are going to use.
We will update our radius by using the following exponential decay function:
$$r(t) = r_0 * e ^ \frac{-t}{c}$$
Where t = the current iteration.<br>
<tab>$r_0$ = The initial value of the map's radius.<br>
c = The update constant that we initialized in our constructor.<br>
    
We can see that we chose in our constructor the following constant $c = \frac{\lambda}{log(r_0)}$ where $\lambda$=max_iter.
If we use this constant in the exponential decay function we get:
$$r(t) = r_0 * e ^ {\frac{-t}{\lambda} * log(r_0)}$$

Given that $log(r_0) \sim ln(r_0)$ we can finally get to:
$$r(t) = r_0 ^ {1 - \frac{t}{\lambda}}$$
This means that as we get through training t gets larger and our radius shrinks exponentially.

```python
class SOM(SOM):
    def receive_iteration_radius(self,iteration):
        '''
        Update our radius using exponential decay function based on the current iteration.
        '''
        return self.radius * math.exp(-iteration / self.update_const)
```

##### Training
As we saw in the algorithm we will use a function $\alpha(t)$ that will change our learning rate over time, this is the rate at which we will update the weights of the nodes at the neighborhood of the given BMU.
Let's implement one training step:. <br>
As we stated in the algorithm's explanation the training phase is based on two important functions:<br>
1. The neighborhood function $\theta(u,v,s)$. <br>
2. The learning rate function $\alpha(s)$. <br>
In the following implementation we will use a method that **combines** these two functions into one function that will produce a matrix that will have the same shape as the SOM grid where the value at index $(i,j)$ in the matrix will describe the learning rate for the node at index $(i,j)$ in the SOM's grid.

```python
class SOM(SOM):
    def training_step(self,trn,num_iter):
        '''
        One training step of our learning algorithm.
        We will compute
        Parameters:
            - trn: The training example that we will train on.
            - num_iter - The current number of iteration.
        '''
        squared_radius = self.receive_iteration_radius(num_iter) ** 2
        
        #Compute the distances of all the points in the grid from the BMU.
        bmu_distances = self.compute_bmu_distances(trn)
        
        #Compute the learning rates matrix.
        learning_rates = np.exp(-bmu_distances / (2 * squared_radius))
        
        return np.expand_dims(learning_rates,2) * (trn - self.weights)
```

##### Explanation 
Let's go over what we did in the training phase (this is the heart of the SOM algorithm).
In line 10 we simply computed the square of the current radius in our map (remember that the radius exponentially decays as we go through the training iterations).<br>
In line 13 we compute the distances of all the nodes on the grid from the best matching unit. Important thing to notice here is that we don't compute the actual distance in the input's dimensional space we only compute the distance between their indices on the grid (for more about this refer to the implementation of compute_bmu_distances)<br>.
In line 16 we compute the learning rates matrix (this is the combination of $\theta(u,v,s)$ and $\alpha(s)$ functions). Formal explanation: If our grid which we will denote as $G$ is of size $(n,m)$ we will denote the matrix of the distances from the BMU as $D$ the final learning rates matrix at iteration t will be defined as:
$$ L[i,j] = e ^ {\frac{- D[i,j]}{2 * r(t) ^ 2}} $$
In the final line of our function (line 18) we simply expand the learning rates matrix to be 3 dimensional and multiply it by the distances of each weights vector of each node from the training example, this will return the amount which we will have to use to update each weight in our weights 3D tensor.

```python
class SOM(SOM):
    def train(self,dataset):
        #Loop until you get to max_iter
        for i in range(self.max_iter):
            #Loop over each training example in the dataset.
            for x in dataset:
                #Update using the function that we defined above.
                self.weights += self.training_step(x,i)
        return self.weights
```

In the function above we simply defined the final training loop, the real logic of the training is in the training_step function so all we have left to implement is the loop which iterates until we get to the maximum iterations and train on each training example using the training_step function.

#### Wait, where is the dimensionality reduction?
When we were first introduced to SOM's we were told that it helps visualizing unlabeled data and is a way to **reduce the dimensionality** of the data. But in our example each node in our map holds a vector of weights that have the same dimensions as the input space. So where is the dimensionality reduction. <br>
You can find a great explanation to this question in the [link](https://stats.stackexchange.com/questions/66304/how-do-soms-reduce-dimensionality-of-data?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa) <br>
**TL;DR**: Basically because we hold all our nodes in 2D grid we can now compute distances between our n-dimensional data on a 2D grid which basically means we reduced it's dimensions from n-dimensions to a 2-dimensional grid.
